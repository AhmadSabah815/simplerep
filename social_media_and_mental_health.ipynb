```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("MentalHealth.csv")
```

```python
# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
# Display the last 5 rows of the DataFrame
display(df.tail())
```

```python
# Display concise summary of the DataFrame
df.info()
```

## Descriptive Statistics

```python
print("### Mean:")
display(df.mean(numeric_only=True))

print("\n### Median:")
display(df.median(numeric_only=True))

print("\n### Mode:")
display(df.mode(numeric_only=True))

print("\n### Standard Deviation:")
display(df.std(numeric_only=True))

print("\n### Variance:")
display(df.var(numeric_only=True))
```

```python
crosstab_gender_social_media = pd.crosstab(df['Gender'], df['Social_Media_Platform'])
display(crosstab_gender_social_media)
```

```python
crosstab_gender_social_media_age = pd.crosstab(index=df['Gender'], columns=df['Social_Media_Platform'], values=df['Age'], aggfunc='mean')
display(crosstab_gender_social_media_age)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Countplot for Gender distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Gender', palette='viridis')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot for Social Media Platform distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Social_Media_Platform', palette='magma', order=df['Social_Media_Platform'].value_counts().index)
plt.title('Distribution of Social Media Platforms')
plt.xlabel('Social Media Platform')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=15, kde=True, color='green')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pivot table for mean Age by Gender and Social Media Platform
pivot_table_age = df.pivot_table(index='Gender', columns='Social_Media_Platform', values='Age', aggfunc='mean')
display(pivot_table_age)
```

```python
# Histogram for Daily Screen Time distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Daily_Screen_Time(hrs)', bins=15, kde=True, color='blue')
plt.title('Distribution of Daily Screen Time (hrs)')
plt.xlabel('Daily Screen Time (hrs)')
plt.ylabel('Frequency')
plt.show()
```
